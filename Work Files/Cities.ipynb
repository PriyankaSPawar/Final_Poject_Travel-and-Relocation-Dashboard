````python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import sqlalchemy
````

##Webscrabe from Wiki information for 15 big EU cities

````python
def cities_dataframe(cities):
  cities_data = []

  for city in cities:
    city_data = {}

    # city
    city_data["city_name"] = city

    # create the soup
    url = f"https://www.wikipedia.org/wiki/{city}"
    response = requests.get(url)
    city_soup = BeautifulSoup(response.content, 'html.parser')

    # country
    city_data["country_name"] = city_soup.find(class_="infobox-data").get_text()

    # population
    city_population = city_soup.find(string="Population").find_next("td").get_text()
    city_population_clean = int(city_population.replace(",", ""))
    city_data["population"] = city_population_clean

    # data retrieved
    city_data["year_data_retrieved"] = city_soup.find(string="Population").find_next().get_text()[2:6]

    # latitude and longitude
    city_data["latitude"] = city_soup.find(class_="latitude").get_text()
    city_data["longitude"] = city_soup.find(class_="longitude").get_text()

    # append this city's data to the cities list
    cities_data.append(city_data)

  return pd.DataFrame(cities_data)
````

````python
cities_df = cities_dataframe(["Berlin", "Madrid", "Paris", "Vienna", "Hamburg", "Bucharest", "Budapest", "Barcelona", "Munich", "Prague", "Milan", "Sofia", "Cologne", "Stockholm", "Amsterdam"])
````

##Creating the database and the backbone of cities, countries & language tables

Before we can send the information in SQL, we need to make tables that have the same columns and data types to recieve the data.

```sql
-- Drop the database if it already exists
-- DROP DATABASE IF EXISTS travel_guide;

-- Create the database
CREATE DATABASE travel_guide;

-- Use the database
USE travel_guide;

CREATE TABLE cities (
    city_id INT AUTO_INCREMENT, -- Automatically generated ID for each city
    city_name VARCHAR(255) NOT NULL, -- Name of the city
    population INT NOT NULL,
    latitude DECIMAL,
    longitude DECIMAL,
    PRIMARY KEY (city_id) -- Primary key to uniquely identify each city
);

CREATE TABLE countries (
    country_id INT AUTO_INCREMENT, -- Automatically generated ID for each country
    country_name VARCHAR(255) NOT NULL -- Name of the country
    city_id INT,
    PRIMARY KEY (country_id),
    FOREIGN KEY (city_id) REFERENCES cities(city_id)
);

CREATE TABLE languages (
    language_id INT AUTO_INCREMENT, -- Automatically generated ID for each language
    language_name VARCHAR(255) NOT NULL -- Name of the language
    PRIMARY KEY (language_id),
    FOREIGN KEY (city_id) REFERENCES cities(city_id)
);
```
