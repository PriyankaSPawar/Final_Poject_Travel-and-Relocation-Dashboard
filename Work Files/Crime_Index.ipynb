```python
import pandas as pd
```

```python
# reading
url = "https://drive.google.com/file/d/1lpJPIKoHJq0HUz0Rfq7LY6Cr1W6tYmDT/view?usp=sharing"
path = 'https://drive.google.com/uc?export=download&id='+url.split('/')[-2]
data = pd.read_csv(path)
```

```python
data.head()
```

```python
# Select the required columns and rename them
selected_columns = {
    'City': 'city_name',
    'Country': 'country_name',
    'Crime Index': 'crime_index',
    'Safety Index': 'safety_index'
}
```

```python
filtered_data = data[selected_columns.keys()].rename(columns=selected_columns)
```

```python
filtered_data
```

```python
# List of selected cities
selected_cities = [
    'Berlin', 'Madrid', 'Paris', 'Vienna', 'Hamburg', 'Bucharest',
    'Budapest', 'Barcelona', 'Munich', 'Prague', 'Milan', 'Sofia',
    'Cologne', 'Stockholm', 'Amsterdam'
]
```

```python
# Filter the DataFrame to include only the selected cities
filtered_data = filtered_data[filtered_data['city_name'].isin(selected_cities)]
```

```python
filtered_data
```

```python
filtered_data.to_csv('./crime_index.csv',index=False)
```
