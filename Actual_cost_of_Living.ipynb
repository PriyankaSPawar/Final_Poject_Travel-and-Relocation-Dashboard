```python
# Columns for each category
food_columns = ['city', 'country', 'Meal_Inexpensive_Restaurant', 'Meal_For_2_Mid_Range_Restaurant', 'McMeal_At_McDonalds',
                'Domestic_Beer_Restaurant', 'Imported_Beer_Restaurant', 'Cappuccino', 'Coke_Pepsi',
                'Water_o.33L', 'Milk_1L', 'Bread_500g', 'Rice_1kg', 'Eggs_12', 'Cheese_1kg', 'Chicken_Fillets_1kg',
                'Beef_1kg', 'Apples_1kg', 'Banana_1kg', 'Oranges_1kg', 'Tomato_1kg', 'Potato_1kg', 'Onion_1kg',
                'Lettuce_1head', 'Water_1.5L_Bottle', 'Bottle_Of_Wine', 'Domestic_Beer_Market', 'Imported_Beer_Market',
                'Cigarettes_20_Pack']
transportation_columns = ['city', 'country', 'One_Way_Ticket_Transport', 'Monthly_Pass_Transport', 'Taxi_Start_Tariff',
                          'Taxi_1km_Tariff', 'Taxi_1hour_Waiting_Tariff', 'Gasoline_1L']
basic_needs_columns = ['city', 'country', 'Basic_Utilities_85m2',
                       'Prepaid_Mobile_Tariff', 'Internet_60Mbps', 'Fitness_Club_Fee', 'Tennis_Court_Rent',
                       'Cinema_1Seat', 'Preschool_Monthly_Fee', 'International_School_Yearly']
housing_columns = ['city', 'country', 'Apartment_1bed_City_Center', 'Apartment_1bed_Outside_Center',
                   'Apartment_3bed_City_Center', 'Apartment_3bed_Outside_Center', 'Price_per_m2_City_Center',
                   'Price_per_m2_Outside_Center']

```

```python
#food_columns
```

```python
# Create DataFrames for each category
food_data = ordered_data[food_columns].copy()
transportation_data = ordered_data[transportation_columns].copy()
basic_needs_data = ordered_data[basic_needs_columns].copy()
housing_data = ordered_data[housing_columns].copy()
```

```python
avg_ex_rate = 0.88  # Replace with the actual average exchange rate
```

```python
# Function to convert USD to Euros based on average exchange rate
def usd_to_eur(price_usd):
    return round(price_usd * avg_ex_rate,2)
```

```python
# Convert USD prices to Euros for each category
for col in food_columns[2:]:  # Start from index 2 to skip 'city' and 'country'
    food_data[col] = food_data[col].apply(usd_to_eur)

for col in transportation_columns[2:]:
    transportation_data[col] = transportation_data[col].apply(usd_to_eur)

for col in basic_needs_columns[2:]:
    basic_needs_data[col] = basic_needs_data[col].apply(usd_to_eur)

for col in housing_columns[2:]:
    housing_data[col] = housing_data[col].apply(usd_to_eur)
```

```python
# Add a 'Category' column to each DataFrame
food_data['Category'] = 'Food'
transportation_data['Category'] = 'Transportation'
basic_needs_data['Category'] = 'Basic Needs'
housing_data['Category'] = 'Housing'
```
